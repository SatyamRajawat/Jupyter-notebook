```python
import os
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
from skimage.transform import resize
```

```python
target = []
images = []
# flat_data = []

categories = []
datadir = 'D:\\Images\\dog_cat'

for category in os.listdir(datadir):
    categories.append(category)

for category in categories:
    path = os.path.join(datadir,category)
    class_num = categories.index(category)

    for img in os.listdir(path):
        img_array = imread(os.path.join(path,img))
        #print(img_array.shape)
        img_resized = resize(img_array,(150,150,3))
#         flat_data.append(img_resized.flatten())
        images.append(img_resized)
        target.append(class_num)
```

```python
# print(flat_data)
# print(categories)
# print(images)
type(target)
```

```python
# flat_data = np.array(flat_data)
target = np.array(target)
images = np.array(images)
```

```python
# print(flat_data.shape)
print(categories)
print(images.shape)
type(target.shape)
```

```python
from sklearn import metrics
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(images,target,test_size=0.2, random_state=242)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
print(y_train[:10])
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.python.keras.utils.data_utils import Sequence
import autokeras as ak
tf.keras.backend.set_floatx('float64')
```

```python
# Initialize the image classifier.

clf = ak.ImageClassifier(
    overwrite=True,
    max_trials=1)
# Feed the image classifier with training data.
history = clf.fit(x_train, y_train, epochs=10)
```

```python
pred_y = clf.predict(x_test)
pred_y
```

```python
from sklearn.metrics import classification, accuracy_score,classification_report, confusion_matrix
```

```python
cm = confusion_matrix(y_test,pred_y)
cm
```

```python
accuracy_score(y_test,pred_y)
```

```python
print(classification_report(y_test,pred_y))
```

```python
print('Accuracy: {accuracy}'.format(accuracy=clf.evaluate(x_test, y_test)))
```
