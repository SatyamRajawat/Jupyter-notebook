```python
#Taking Independent variable on new dataframe
df1 = df[['satisfaction_level','average_montly_hours','promotion_last_5years','Work_accident','Department','salary']]
df1.head()
```

# Order encoder for salary

```python
# Using order encoder to convert salary 
df1['salary'].value_counts()
```

```python
order_label = {'low':1,'medium':2,'high':3}
```

```python
df1['salary_ord'] = df1['salary'].map(order_label)
df1.sample(10)
```

```python
df1.shape
```

```python
# One Hot encoding for Department
df1['Department'].value_counts()
```

```python
# df2 = ohe.fit_transform(df1['Department'])
df2 = pd.get_dummies(df1.Department,drop_first =True)
df2.head()
```

```python
df2.shape
```

```python
df3 = pd.concat([df1, df2],axis='columns')
df3.head()
```

```python
df3.isnull().sum()
```

```python
df_final = df3.drop(['Department','salary'],axis=1)
df_final.sample(10)
```

```python
X = df_final
X.head()
```

```python
y = df['left']
y.head()
```

# Training our model

```python
# Spliting data into train Test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```

```python
import numpy as np
def predict_employee(Department,satisfaction_level,average_montly_hours,promotion_last_5years,Work_accident,salary_ord):    
    loc_index = np.where(X.columns==Department)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = satisfaction_level
    x[1] = average_montly_hours
    x[2] = promotion_last_5years
    x[3] = Work_accident
    x[4] = salary_ord
    if loc_index >= 0:
        x[loc_index] = 1
    return model.predict([x])[0]
```

```python
# mod.predict([['Department','satisfaction_level','average_montly_hours','promotion_last_5years','salary_ord']])
predict_employee('sales',0.15,270,0,0,2)
```
